```python
# Tentukan direktori dataset
dataset_dir = 'fruits_dataset/fruits-360_dataset_original-size'

# Inisialisasi dictionary untuk menyimpan resolusi
resolutions = {}

# Fungsi untuk memeriksa resolusi gambar
def check_image_resolutions(directory):
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith('.jpg') or file.endswith('.png'):
                file_path = os.path.join(root, file)
                with Image.open(file_path) as img:
                    resolution = img.size  # (width, height)
                    if resolution in resolutions:
                        resolutions[resolution] += 1
                    else:
                        resolutions[resolution] = 1

# Cek resolusi gambar dalam dataset
check_image_resolutions(dataset_dir)

# Tampilkan hasil
for resolution, count in resolutions.items():
    print(f'Resolution: {resolution}, Count: {count}')
```

```python
# Path Dataset
dataset_dir = 'fruits_dataset/fruits-360_dataset_original-size/fruits-360-original-size'

# Folder yang akan digabungkan jadi 1
subfolders = ['Training', 'Test', 'Validation']

# Membuat folder gabungan
combined_dir = 'fruits_dataset/combined'

# Bikin folder jika belum ada
os.makedirs(combined_dir, exist_ok=True)

# Fungsi untuk menggabungkan file dari subfolder sambil mempertahankan struktur subfolder berdasarkan label
def combine_files_with_structure(directory, subfolders, combined_directory):
    for subfolder in subfolders:
        subfolder_path = os.path.join(directory, subfolder)
        for root, dirs, files in os.walk(subfolder_path):
            for file in files:
                if file.endswith('.jpg') or file.endswith('.png'):
                    # Mempertahankan struktur isi folder
                    relative_path = os.path.relpath(root, subfolder_path)
                    destination_subfolder_path = os.path.join(combined_directory, relative_path)
                    os.makedirs(destination_subfolder_path, exist_ok=True)

                    # Tujuan file yang akan dicopas
                    source_file_path = os.path.join(root, file)
                    destination_file_path = os.path.join(destination_subfolder_path, file)

                    # Copas file
                    shutil.copy2(source_file_path, destination_file_path)

combine_files_with_structure(dataset_dir, subfolders, combined_dir)
```

## Augmentasi Gambar pada Data Train

```python
# Inisialisasi ImageDataGenerator dengan augmentasi
datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Fungsi untuk augmentasi gambar
def augment_images(data_dir, augment_dir, num_augmented_images=5, target_size=(224, 224)):
    os.makedirs(augment_dir, exist_ok=True)
    for class_name in os.listdir(data_dir):
        class_dir = os.path.join(data_dir, class_name)
        augment_class_dir = os.path.join(augment_dir, class_name)
        os.makedirs(augment_class_dir, exist_ok=True)
        for img_name in os.listdir(class_dir):
            img_path = os.path.join(class_dir, img_name)
            img = load_img(img_path, target_size=target_size)
            x = img_to_array(img)
            x = x.reshape((1,) + x.shape)
            i = 0
            for batch in datagen.flow(x, batch_size=1, save_to_dir=augment_class_dir, save_prefix='aug', save_format='jpg'):
                i += 1
                if i >= num_augmented_images:
                    break

# Augmentasi gambar dalam folder train dan pisahkan ke folder augmentasi
augment_dir = 'Dataset-Final/augmented_train'
augment_images(train_dir, augment_dir)

print("=====================")
print("Done bang")
print("=====================")
```

```python
# Path dataset
train_dir = 'Dataset-Final/train'
augment_dir = 'Dataset-Final/augmented_train'
test_dir = 'Dataset-Final/test'


# Fungsi untuk memindahkan file dari folder augmentasi ke folder train
def move_augmented_files(augment_dir, train_dir):
    for class_name in os.listdir(augment_dir):
        augment_class_dir = os.path.join(augment_dir, class_name)
        train_class_dir = os.path.join(train_dir, class_name)
        os.makedirs(train_class_dir, exist_ok=True)
        for img_name in os.listdir(augment_class_dir):
            src_path = os.path.join(augment_class_dir, img_name)
            dst_path = os.path.join(train_class_dir, img_name)
            shutil.move(src_path, dst_path)

move_augmented_files(augment_dir, train_dir)

# Fungsi untuk menghitung jumlah gambar setiap kelas
def count_files_in_directory(dir):
    counts = {}
    for class_name in os.listdir(dir):
        class_path = os.path.join(dir, class_name)
        if os.path.isdir(class_path):
            counts[class_name] = len(os.listdir(class_path))
    return counts

train_counts = count_files_in_directory(train_dir)

# Menampilkan jumlah file dalam setiap kelas
for label, count in train_counts.items():
    print(f"{label}: {count} gambar")
```

```python
train_dir = 'Dataset-Final/train'
augment_dir = 'Dataset-Final/augmented_train'
test_dir = 'Dataset-Final/test'
```

```python
# Path dataset untuk ImageDataGenerator
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Buat generator gambar
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=64,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=64,
    class_mode='categorical'
)
```

## Modeling dengan PreTrained VGG16

```python
# Load model VGG16 tanpa top layer
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
model = Sequential()
model.add(base_model) # dari VGG16

# Setup layer
model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.3))

model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.3))

model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(len(train_generator.class_indices), activation='softmax'))
model.summary()

# Freeze base_model
for layer in base_model.layers:
    layer.trainable = False

# Compile
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Setup Callbacks
checkpoint_path = 'Dataset-Final/model_checkpoint.h5'
checkpoint = ModelCheckpoint(checkpoint_path, monitor='val_loss', save_best_only=True, mode='min', verbose=1)
early_stopping = EarlyStopping(monitor='val_loss', patience=3, mode='min', verbose=1)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.001)

callbacks = [checkpoint, early_stopping, reduce_lr]

history = model.fit(
    train_generator,
    epochs=5, # Melihat epoch permulaan sudah dapat akurasi tinggi
    validation_data=test_generator,
    callbacks=callbacks
)
```

```python
for layer in base_model.layers[-4:]:
    layer.trainable = True

# Compile ulang dengan Adam yang di Custom
model.compile(optimizer=Adam(learning_rate=1e-5),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history_new = model.fit(
    train_generator,
    epochs=5,
    validation_data=test_generator,
    callbacks=callbacks
)
```

## Evaluasi dan Visualisasi Akurasi dan Loss

```python
# Evaluasi akurasi pada data train
train_loss, train_accuracy = model.evaluate(train_generator)
print(f"Akurasi pada data Train: {train_accuracy * 100:.2f}%")
print(f"Train Loss: {train_loss}")

# Evaluasi akurasi pada data test atau val
test_loss, test_accuracy = model.evaluate(test_generator)
print(f"Akurasi pada data Val: {test_accuracy * 100:.2f}%")
print(f"Val Loss: {test_loss}")
```
